```python
# Change BASE_DIR to point to the extracted directory
BASE_DIR = '/content/downsampled'  # Point to the extracted directory
age_labels = []
gender_labels = []
image_paths = []


image_filenames = os.listdir(BASE_DIR)
random.shuffle(image_filenames)

for image in tqdm(image_filenames):
  image_path = os.path.join(BASE_DIR, image)
  img_components = image.split('_')
  age_label = int(img_components[0])

  # Handle potential non-integer gender labels
  try:
      gender_label = int(img_components[1])
  except ValueError:
      gender_label = None  # Or any other handling you prefer for non-integer labels
      print(f"Non-integer gender label encountered in filename: {image}")

  # Append the image_path, age_label, and gender_label
  age_labels.append(age_label)
  gender_labels.append(gender_label)
  image_paths.append(image_path)
```

```python
print(f'Number of age_labels: {len(age_labels)}, Number of gender_labels: {len(gender_labels)}, Number of image_paths: {len(image_paths)}')
```

```python
gender_mapping = {
    1: 'Female',
    0: 'Male'
}
```

```python
import pandas as pd
df = pd.DataFrame()
df['image_path'], df['age'], df['gender'] = image_paths, age_labels, gender_labels
df.head(5)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import random

rand_index = random.randint(0, len(image_paths))
age = df['age'][rand_index]
gender = df['gender'][rand_index]
IMG = Image.open(df['image_path'][rand_index])

# Handle the case where gender is NaN
if pd.isna(gender):
    gender_string = "Unknown"  # Or any other string you prefer
else:
    gender_string = gender_mapping.get(int(gender), "Unknown")  # Handle potential non-existing keys

plt.title(f'Age: {age} Gender: {gender_string}')
plt.axis('off')
plt.imshow(IMG)
plt.show()  # Add this line to actually display the image
```
